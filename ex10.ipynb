## Тема 8 “Понятие об интеграле”



**1. Найти неопределенный интеграл**

$ \large{ \int (2x^2-2x-1+sin(x)-cos(x)+ln(x)+e^x)dx } $

Интегрируем почленно в соответствии с основными интегралами

$ \large{ \int 2x^2dx= \frac{2x^3}{3} } $

$ \large{ \int -2xdx= -x^2 } $

$ \large{ \int -1dx= -x } $

$ \large{ \int sin(x)dx= -cos(x) } $

$ \large{ \int cos(x)dx= sin(x) } $

$ \large{ \int e^xdx= e^x } $

Ответ:


$ \large{ \frac{2x^3}{3}-x^2+xln(x)-2x+e^x-\sqrt{2}sin(x+\frac{\pi}{4})+constant } $

___


**2. Найти неопределенный интеграл**

$ \large{ \int (2x+6xz^2-5x^2y-3ln(z))dx } $

Интегрируем почленно в соответствии с основными интегралами

$ \large{ \int (2x)dx = x^2 } $

$ \large{ \int (6xz^2)dx = 3x^2z^2 } $

$ \large{ \int (-5x^2y)dx = \frac{-5x^2y}{3} } $

$ \large{ \int (3ln(z))dx = -3xln(z) } $

Ответ:

$ \large { -\frac{5x^3y}{3} + 3x^2z^2+x^2-3xln(z) } $

$ \large{ = \frac{x(-5x^2y+9xz^2+3x-9ln(z))}{3} } $

$ \large{ = -\frac{5y}{3}+3z^2-3ln(z)+1 } $

**3. Найти определенный интеграл**

$\large { \int_{0}^{\pi} (3x^2sin(2x))dx }$

$ \large { 3 \int_{0}^{\pi} x^2sin(2x)dx } $

$ = \large { -\frac{3}{2}x^2cos(2x)+ 3 \int_{0}^{\pi} xcos(2x)dx } $

Подставим пределы интеграла

$ = \large { -\frac{3\pi^2}{2} + 3 \int_{0}^{\pi} xcos(2x)dx } $

$ = \large { -\frac{3\pi^2}{2} + \frac{3}{2}xsin(2x) - \frac{3}{2} \int_{0}^{\pi} sin(2x)dx } $

Подставим пределы интеграла

$ = \large { -\frac{3\pi^2}{2} - \frac{3}{2} \int_{0}^{\pi} sin(2x)dx } $

$ = \large { -\frac{3\pi^2}{2} - \frac{3}{4} \int_{0}^{2\pi} sin(u)dx } $

Подставим пределы интеграла

Ответ:

$ = \large { -\frac{3\pi^2}{2} } $


**4. Найти неопределенный интеграл**

$ \large{ \int (\frac{1}{\sqrt{x+1}})dx } $

Заменим $ \large { u = \sqrt{x+1} } $

$ \large{ du = \frac{dx}{2\sqrt{x+1}} = 2u } $

Ответ:

$ \large{ 2\sqrt{x+1} + constant } $
