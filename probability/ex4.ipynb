**3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения**

$ \large{ f(x) = \frac{1}{(4\sqrt{2\pi}))} * (\frac{exp(-(x+2)^2)}{32}) } $

**Найдите:**

- а). M(X)
- б). D(X)
- в). std(X) (среднее квадратичное отклонение)

Возьмем пределы:

$ \large{ 0 < x < 1 } $

Мат. ожидание:

$ \large{ M[x] = \int_{0}^{1}x* (\frac{1}{(4\sqrt{2\pi}))} * (\frac{exp(-(x+2)^2)}{32}))dx = (\frac{-2\sqrt{2}*x-\sqrt{2}*e^{-2*x-4}}{1024\sqrt{\pi}}) = \frac{-2\sqrt{2}*1-\sqrt{2}*e^{-2*1-4}}{1024\sqrt{\pi}} - \frac{-2\sqrt{2}*0-\sqrt{2}*e^{-2*0-4}}{1024\sqrt{\pi}} = 0.0 }$

Дисперсия:

$ \large{ D[x] = \int_{0}^{1}x^2 * (\frac{1}{(4\sqrt{2\pi}))} * (\frac{exp(-(x+2)^2)}{32}))dx - 0.0^2 = 0.0}$

Среднеквадратическое отконение:

$ \large{ s = \sqrt{D[x]} = \sqrt{0.0} = 0 } $
