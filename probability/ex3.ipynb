**1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean)**
- среднее арифметическое, 
- среднее квадратичное отклонение, 
- смещенную и несмещенную оценки дисперсий 

**для данной выборки.**

Среднее арифметическое:

$ \large{ M(X) = \frac{1}{n} \sum\limits_{i=1}^{n} x_i = \frac{1315}{20} = 65.75 }$

Cреднее квадратичное отклонение:

$ \large{ \sigma = \sqrt{\frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}} = \sqrt{\frac{19343.75}{20}} = 31.099 } $

Смещенная оценка дисперсии:

$ \large{ \sigma^2 = \frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n} } = \frac{19343.75}{20} = 967.188 $

Несмещенная оценка дисперсии:

$ \large{ \sigma^2_{несмещ.} = \frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n - 1} = \frac{19343.75}{20-1} = 1018.092 } $

Ниже проверим все в коде:

```python
import numpy as np
from math import factorial
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
items = np.array([100,89,75,77,89,33,45,25,65,17,30,24,57,55,70,75,65,84,90,150])
df = pd.DataFrame(items,columns=['a'])

mean = df['a'].sum() / df['a'].count()
print(f'Среднее арифметическое: {mean}')

std = np.sqrt(((df['a'] - df['a'].mean())**2).sum() / df['a'].count())
print(f'Среднее квадратическое отклонение: {std}')

variance = ((df['a'] - df['a'].mean())**2).sum() / df['a'].count()
print(f'Смещенная оценка дисперсии: {variance}')

variance2 = ((df['a'] - df['a'].mean())**2).sum() / (df['a'].count() - 1)
print(f'Несмещенная оценка дисперсии: {variance2}')
```

**2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?**

Комбинации:
- в первом 0 во втором 3
- в первом 1 во втором 2
- в первом 2 во втором 1

$ \large{ \frac{C_{5}^0*C_{3}^2}{C_{8}^2} * \frac{C_{5}^3*C_{8}^0}{C_{12}^4} + \frac{C_{5}^1*C_{3}^1}{C_{8}^2} * \frac{C_{5}^2*C_{8}^2}{C_{12}^4} + \frac{C_{5}^2*C_{3}^0}{C_{8}^2} * \frac{C_{5}^1*C_{8}^3}{C_{12}^4} = 0.507 } $

```python
comb1 = (combinations(5,0)*combinations(3,2)/combinations(8,2)) * (combinations(5,3)*combinations(8,0)/combinations(12,4))
comb2 = (combinations(5,1)*combinations(3,1)/combinations(8,2)) * (combinations(5,2)*combinations(8,2)/combinations(12,4))
comb3 = (combinations(5,2)*combinations(3,0)/combinations(8,2)) * (combinations(5,1)*combinations(8,3)/combinations(12,4))
res = comb1 + comb2 + comb3
print(f'{comb1} + {comb2} + {comb3} = {res}')
```

**3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.**

3 гипотезы:
- стреляет первый стрелок
- стреляет второй стрелок
- стреляет третий стрелок

Вероятность того что стреляет любой стрелок = $ \large{ \frac{1}{3} } $

Вероятность выстрела первым стрелком:

$ \large{P(A,B) = \frac{0.9*\frac{1}{3}}{0.9*\frac{1}{3}+0.8*\frac{1}{3}+0.6*\frac{1}{3}} = 0.391 } $

Вероятность выстрела вторым стрелком:

$ \large{P(A,B) = \frac{0.8*\frac{1}{3}}{0.9*\frac{1}{3}+0.8*\frac{1}{3}+0.6*\frac{1}{3}} = 0.347 } $

Вероятность выстрела третим стрелком:

$ \large{P(A,B) = \frac{0.6*\frac{1}{3}}{0.9*\frac{1}{3}+0.8*\frac{1}{3}+0.6*\frac{1}{3}} = 0.260 } $

```python
a1 = (0.9*(1/3))
a2 = (0.8*(1/3))
a3 = (0.6*(1/3))

b = (0.9*(1/3))+(0.8*(1/3))+(0.6*(1/3))

print(f'Вероятнорсть выстрела первым стрелком: {a1/b}')
print(f'Вероятнорсть выстрела вторым стрелком: {a2/b}')
print(f'Вероятнорсть выстрела третим стрелком: {a3/b}')
```

- а). все детали 
- б). только две детали 
- в). хотя бы одна деталь 
- г). от одной до двух деталей?

Все детали:

$ P(A) = 0.1*0.2*0.25 = 0.005 $

$ 0.1*0.2*(1-0.25) + (1-0.1)*0.2*0.25 + 0.1*(1-0.2)*0.25 = 0.80 $
