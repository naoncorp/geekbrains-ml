$ \large { 144 = 2(x+y)} $

$ \large { 72 = x+y } $

$ \large { 72 / 2 = 36 } $

```python
rg = range(1, 37)
v = []
res = []

for x in rg:
    for y in reversed(rg):
        v.append({x, y})
        res.append(x*y)
    
#print(res)
print(max(res))
#print(v)
print(v[res.index(max(res))])
```
