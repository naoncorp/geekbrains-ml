## 1. Исследовать на линейную зависимость:

$ f_1(x)=e^x, f_2(x)=1, f_3(x)=x+1, f_4(x)=x-e^x $

**Решение:**

$ f_4 = f_3-f_2-f_1  $, т.е. $ (x-e^x) = (x+1) - 1 - (e^x) $

Т.к.один из векторов является линейной комбинацией двух остальных - векторы линейно зависимы.

## 2. Исследовать на линейную зависимость:

$ f_4 = f_3+2f_2+\frac{f_1}{2}  $, т.е. $ (x+1)^{2} = x^{2} + 2x + \frac{2}{2} $

Т.к.один из векторов является линейной комбинацией двух остальных - векторы линейно зависимы.

## 1. Найти скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

Для расчета понадобится сложить произведения значений векторов.

```python
import numpy as np
import math
from numpy import linalg as LA

x = np.array([0,-3,6])
y = np.array([-4,7,9])
res = np.dot(x,y)
print(f"Ответ а: Скалярное произведение x и y = {res}")

x = np.array([7, -4, 0, 1])
y = np.array([-3, 1, 11, 2])
res = np.dot(x,y)
print(f"Ответ б: Скалярное произведение x и y = {res}")
```

## 2. Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними:

$ L_1 = |4+2+4| = 10 $

$ L_1 = |12+3+4| = 19 $

$ L_2 = \sqrt{4^2+2^2+4^2} = 6 $

$ L_2 = \sqrt{12^2+3^2+4^2} = 13 $

Скалярное произведение = $ 70 $

Угол между векторами = $ arccos(70/(13*6)) = 0.456 $

```python
x = np.array([4,2,4])
y = np.array([12,3,4])

print(f"Манхетовская норма вектора x: {LA.norm(x, 1)}")
print(f"Евклидова норма вектора x: {LA.norm(x, 2)}")
print(f"Манхетовская норма вектора y: {LA.norm(y, 1)}")
print(f"Евклидова норма вектора y: {LA.norm(y, 2)}")

angle = math.acos(np.dot(x,y)/(LA.norm(x, 2)*LA.norm(y, 2)))
print(f"Угол между векторами x и y: {angle}")
```

## 3. Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
**а) произведение длин векторов;**<br>

Для решения этой задачи необходимо проверить выполнение аксиом. В данном случае не выполняется следующая аксиома:

$ (x_1+x_2), y = (x_1,y)+(x_2,y) $

Проверим это на конкретном примере:

$ x_1=(8,0), x_2 = (0,4), y = (7,0) $

$ |x_1| = \sqrt{8_2+0^2} = \sqrt{64} = 8 $

$ |x_2| = \sqrt{0_2+4^2} = \sqrt{16} = 4 $

$ x_1 + x_2 = (8,4) $

$ |x_1+x_2| = \sqrt{8^2+4^2} = \sqrt{80} = 8.944 $

$ |y| = 7 $

Проверим равенство:

$ |x_1+x_2| * y = |x_1|*|y| + |x_2|*|y| $

$ 8.944*7 != 8*7 + 4*7 $

**Ответ** Пространство не будет евклидовым


**б) утроенное обычное скалярное произведение векторов?**<br>

Пространство будет евклидовым.

## 4. Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ - Не образуют базис, так как не хватает одного вектора в R^3 <br>

б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>

Ответ: Образуют, т.к. Евклидова норма = 0

```python
import numpy as np
import math
from numpy import linalg as LA
a = np.array([1/math.sqrt(2), -1/math.sqrt(2), 0])
b = np.array([1/math.sqrt(2), 1/math.sqrt(2), 0])
c = np.array([0, 0, 1])
print(f"Скалярное произведение: ab={np.dot(a,b)}, ac={np.dot(a,c)}, bc={np.dot(b,c)}")
print(f"Евклидова норма: a={LA.norm(a)}, b={LA.norm(b)}, c={LA.norm(c)}")
```
