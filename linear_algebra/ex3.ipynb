## 1. Найти собственные векторы и значения для линейного оператора, заданного матрицей

$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}$

Составляем систему для определения координат собственных векторов:

$(-1 - λ)x_1-6x_2 = 0 $

$ 2x_1 + (6 - λ)x_2 = 0 $

Найдем собственные значения. Подставим $\lambda$ и вычислим определитель:

$\begin{vmatrix}
-1-\lambda & -6\\ 
2 & 6-\lambda
\end{vmatrix} = \lambda^2-5\lambda+6 = (\lambda-2)*(\lambda-3)$

$\lambda_1 = 2$

$\lambda_2 = 3$

$ -4x_1-6y_1 = 0 $

$ 2x_1 + 3y_1 = 0 $

Найдем собственные вектора:

Собственный вектор, отвечающий числу $λ_1$ = 3 при $x_1$ = 3:

$x_1 = (3,-2)$

В качестве единичного собственного вектора принимаем вектор:

$ \large{ i_1 = (\frac{3}{\sqrt{13}};\frac{-2}{\sqrt{13}}) }$

Координаты второго собственного вектора, соответствующего второму собственному числу $λ_2 = 2$, находим из системы:

$ -3x_1-6y_1 = 0 $

$ 2x_1 + 4y_1 = 0 $

$ x_2 = (2,-1) $

$ \large{ j_1 = (\frac{2}{\sqrt{2}};\frac{-1}{\sqrt{2}})} $




## 2 Дан оператор поворота на 180 градусов, задаваемый матрицей 

$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$

Показать, что __любой__ вектор является для него собственным.

Найдем собственные числа:

$ \begin{vmatrix}
-1-\lambda & 0\\ 
0 & -1-\lambda
\end{vmatrix} = \lambda^2+2\lambda+1 = (\lambda+1)*(\lambda+1) $

$ \lambda = -1 $

$ \begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} = \lambda \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix}  $

Любой вектор подходит


## 3. Пусть линейный оператор задан матрицей

$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} = \lambda\begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} $

Подставим значения в систему уравнений:

$
\begin{cases}
x_1-x_2=\lambda x_1\\
-x_1+3x_2 = \lambda x_2
\end{cases} = \begin{cases}
\lambda = 2\\
\lambda = 2
\end{cases}$

Вектор x = (1,1) является собственным вектором
