**Задание 1:**

$ \large{ U = 3-8x+6y } $ если $ \large{ U = x^2+y^2=36 } $

$ \large{ L(\lambda_1,x,y) = 3-8x+6y+\lambda_1*(x^2+y^2-36) } $

Находим производные в системе уравнений

$ \large{ \begin{equation*}
 \begin{cases}
   L_x' = -8 + \lambda_1 * 2x = 0
   \\
   L_y' = 6 + \lambda_1 * 2y = 0
   \\
   L_{\lambda_1}' = x^2 + y^2 - 36 = 0
 \end{cases}
\end{equation*} } $

$ \large{ \begin{equation*}
 \begin{cases}
   x = \frac{4}{\lambda_1}
   \\
   y = -\frac{3}{\lambda_1}
   \\
   \lambda_1^2 = \frac{25}{36}
 \end{cases}
\end{equation*} } $

Ответ:

$ \large{ (\frac{5}{6},\frac{24}{5}, -\frac{18}{5}), (-\frac{5}{6},-\frac{24}{5}, \frac{18}{5}) } $

Исследуем достаточное условие

$ \large{ L_x' = -8 + \lambda_1 * 2x } $

$ \large{ L_y' = 6 + \lambda_1 * 2y } $

$ \large{ L_{\lambda_1}' = x^2 + y^2 - 36 } $

Найдем вторые производные

$ \large{ L_{xx}'' = 2\lambda_1 } $

$ \large{ L_{yy}'' = 2\lambda_1 } $

$ \large{ L_{\lambda_1\lambda_1}'' = 0 } $

$ \large{ L_{xy}'' = 0 } $

$ \large{ L_{x\lambda_1}'' = 2x } $

$ \large{ L_{y\lambda_1}'' = 2y } $

Найдем определитель

$\large{
\begin{pmatrix}
L_{\lambda_1\lambda_1}'' & L_{x\lambda_1}'' & L_{y\lambda_1}''\\
L_{x\lambda_1}'' & L_{xx}'' & L_{xy}''\\
L_{y\lambda_1}'' & L_{xy}'' & L_{yy}''
\end{pmatrix}}$

$\large{
= \begin{pmatrix}
0 & 2x & 2y\\
2x & 2\lambda_1 & 0\\
2Y & 0 & 2\lambda_1
\end{pmatrix}}$

$\large{
= \begin{vmatrix}
0 & 2x & 2y\\
2x & 2\lambda_1 & 0\\
2Y & 0 & 2\lambda_1
\end{vmatrix}}$

$ \large{ = 0-2x*(2x*2\lambda_1-0)+2y*(0-2y*2\lambda_1) } $

$ \large{ = -8x^2\lambda_1-8y^2\lambda_1 = -8\lambda_1(x^2+y^2) } $

$ \large{ (\frac{5}{6},\frac{24}{5}, -\frac{18}{5}), (-\frac{5}{6},-\frac{24}{5}, \frac{18}{5}) } $

## Найти производную функции в точке по направлению вектора

Формула

$ \large{ U_{\vec{c}}' = (\vec{c_0*gradU}) } $

**Задание 1**
Найти производную функции $ U=x^2+y^2+z^2 $ по направлению вектора $ \vec{c}(-9,8,-12) $ в точке $ M(8,-12,9) $

Решение:

$ \large{ |\vec{c}| = \sqrt{x_0^2+y_0^2+z_0^2} = \sqrt{ (-9)^2 +8^2 + (-12)^2 } = \sqrt{289} = 17 } $

$ \large{ \vec{c_0} = \frac{\vec{c}}{|\vec{c}|} = (-\frac{9}{17},\frac{8}{17},-\frac{12}{17}) } $

Найдем производные

$ \large{ gradU=(2x,2y,2z) } $

$ \large{ gradU|_m = (16,-24,18) } $

$ \large{ U_{\vec{c}}'|_m = -\frac{9}{17}*1+\frac{8}{17}*(-24)-\frac{12}{17}*18 = -\frac{552}{17} } $

**Задание 2**

Найти производную функции $ U=e^{x^2+y^2+z^2} $ по направлению вектора $ \vec{d}(4,-13,-16) $ в точке $ L(-16,4,-13) $

$ \large{ |\vec{d}| = \sqrt{x_0^2+y_0^2+z_0^2} = \sqrt{ (4)^2 +(-13)^2 + (-16)^2 } = \sqrt{441} = 21 } $

$ \large{ \vec{d_0} = \frac{\vec{d}}{|\vec{d}|} = (\frac{4}{21},-\frac{13}{21},-\frac{16}{21}) } $


Найдем производные

$ \large{ gradU=(e^{x^2+y^2+z^2}*2x,e^{x^2+y^2+z^2}*2y,e^{x^2+y^2+z^2}*2z) } $

$ \large{ gradU|_l = (-32e^{441},8e^{441},-26e^{441}) } $

$ \large{ U_{\vec{d}}'|_l = \frac{4}{21}*(-32e^{441})-\frac{13}{21}*(8e^{441})-\frac{16}{21}*(-26e^{441}) = -\frac{184}{21}e^{441} } $
